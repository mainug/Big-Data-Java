```python
a = "banana"
b = a
print(id(a))
print(id(b)) # a와 b의 주소값이 다름
b = "kiwi"
print(id(a)) # a와 b의 주소값이 다름
print(id(b)) # a와 b의 주소값이 다름
```

```python
aa = [11, 22, 33]
print(id(aa))
bb = aa
print(id(bb))
bb[0] = 33
print(id(aa))
print(id(bb))
# 리스트 객체가 메모리에 생성되고 aa는 그 객체를 참조
# bb = aa 는 새로운 리스트를 만드는 것이 아닌 aa가 참조하는 객체를
# bb도 참조하기에 이런 결과가 나온 것
# 같은 리스트를 참조하는 다른 이름인 것
```

```python
aa = [11, 22, 33]
# bb = aa.copy()
# bb = list(aa)
bb = aa[:]
print(id(aa))
print(id(bb))
```

### 리스트 관련 함수

```python
# append - 추가하기
a = []
a.append(1)
print(a)
a.append(2)
print(a)
a.append([11, 22, 33])
print(a)
a.append(4)
print(a)
```

```python
# sort - 리스트 정렬
a = [5 ,24 ,75, 2, 43, 36 ,93]
b = ['gf', 'f', 'u', 'a', 'e', 'f', 'z', 'c']
a.sort()
b.sort()
print(a)
print(b)
```

```python
# reverse - 리스트 뒤집기
a = [5 ,24 ,75, 2, 43, 36 ,93]
b = ['gf', 'f', 'u', 'a', 'e', 'f', 'z', 'c']
# 원본 훼손 X
print(a[::-1])
print(b[::-1])
# 원본 훼손
a.reverse()
b.reverse()
print(a)
print(b)
```

```python
# sort와 reverse 응용
a = [5 ,24 ,75, 2, 43, 36 ,93]
a.sort()
a.reverse() # a[::-1]
print(a)

b = [5 ,24 ,75, 2, 43, 36 ,93]
b.sort(reverse=True) # True는 내림차순, False는 오름차순이 기본값
print(b)

# sort, reverse는 반환값 없음(결과: None)
print(a.sort())
print(a.reverse())
```
